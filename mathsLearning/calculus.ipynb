```python
import pandas as pd
import torch
import pickle
import matplotlib.pyplot as plt

# 📥 Load CSV data
df = pd.read_csv("house_prices.csv")  # Replace with your actual path

# 🧹 Preprocess
X_np = df.drop(columns=["price"]).values
y_np = df["price"].values

# Convert to tensors
X = torch.tensor(X_np, dtype=torch.float32)
y = torch.tensor(y_np, dtype=torch.float32).view(-1, 1)

# 🔢 Initialize weights and bias manually
n_features = X.shape[1]
W = torch.randn(n_features, 1, requires_grad=True)
b = torch.randn(1, requires_grad=True)

# 🧮 Custom MSE loss function
def mse(pred, target):
    return torch.mean((pred - target) ** 2)

# 📊 Plotting function
def plot_fit(epoch, X, y, W, b):
    if X.shape[1] != 1:
        print(f"Skipping plot at epoch {epoch}: only works for 1 feature.")
        return

    with torch.no_grad():
        x_vals = X[:, 0].numpy()
        y_vals = y.numpy().flatten()
        y_pred = (X @ W + b).numpy().flatten()

        plt.figure(figsize=(8, 5))
        plt.scatter(x_vals, y_vals, label="Actual", alpha=0.6)
        plt.plot(x_vals, y_pred, color="red", label="Predicted Line")
        plt.title(f"Fitment at Epoch {epoch}")
        plt.xlabel("Feature")
        plt.ylabel("Price")
        plt.legend()
        plt.grid(True)
        plt.show()

# 🔁 Training loop
lr = 0.01
epochs = 10000

for epoch in range(epochs):
    y_pred = X @ W + b
    loss = mse(y_pred, y)

    loss.backward()

    with torch.no_grad():
        W -= lr * W.grad
        b -= lr * b.grad
        W.grad.zero_()
        b.grad.zero_()

    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")
        plot_fit(epoch, X, y, W, b)

# 💾 Save weights and bias to pickle
model_params = {"W": W.detach(), "b": b.detach()}
with open("manual_house_price_model.pkl", "wb") as f:
    pickle.dump(model_params, f)

print("✅ Model trained and saved to 'manual_house_price_model.pkl'")
```

```python
import os
import fnmatch

def load_ignore_patterns(ignore_file_path):
    """
    Loads exclusion patterns from a file, ignoring comments and empty lines.
    """
    patterns = []
    with open(ignore_file_path, 'r') as f:
        for line in f:
            stripped_line = line.strip()
            if stripped_line and not stripped_line.startswith('#'):
                patterns.append(stripped_line)
    return patterns

def is_ignored(item_name, item_path, ignore_patterns):
    """
    Checks if a file or directory should be ignored based on the patterns.
    """
    for pattern in ignore_patterns:
        if pattern.endswith('/'):
            if os.path.isdir(item_path) and fnmatch.fnmatch(item_name, pattern.rstrip('/')):
                return True
        else:
            if fnmatch.fnmatch(item_name, pattern):
                return True
    return False

def display_directory_tree(root_dir, ignore_file_path=None):
    """
    Prints a visual tree structure of a directory, respecting ignore patterns.
    """
    if not os.path.isdir(root_dir):
        print(f"❌ Error: The provided path '{root_dir}' is not a valid directory.")
        return

    print("--- 🕵️‍♂️ Debug Info ---")
    ignore_patterns = []
    
    # Logic to find and load the ignore file
    if ignore_file_path and os.path.isfile(ignore_file_path):
        print(f"Using user-provided ignore file: {ignore_file_path}")
        ignore_patterns = load_ignore_patterns(ignore_file_path)
    else:
        default_ignore_file = os.path.join(root_dir, '.gitignore')
        print(f"Searching for default ignore file at: {default_ignore_file}")
        if os.path.isfile(default_ignore_file):
            print("✅ Found .gitignore in root directory.")
            ignore_patterns = load_ignore_patterns(default_ignore_file)
        else:
            print("❌ No explicit or default .gitignore file found.")

    if ignore_patterns:
        print(f"📜 Patterns loaded: {ignore_patterns}")
    else:
        print("📜 No ignore patterns were loaded. Listing all files.")
    
    print("----------------------")
    
    print(f"\n📁 {os.path.basename(root_dir)}")
    _generate_tree(root_dir, prefix="", ignore_patterns=ignore_patterns)

def _generate_tree(directory, prefix="", ignore_patterns=None):
    if ignore_patterns is None:
        ignore_patterns = []

    try:
        items = sorted(os.listdir(directory))
    except PermissionError:
        print(prefix + "└── [Permission Denied]")
        return
    
    filtered_items = []
    for item_name in items:
        item_path = os.path.join(directory, item_name)
        if not is_ignored(item_name, item_path, ignore_patterns):
            filtered_items.append(item_name)
            
    pointers = ['├── '] * (len(filtered_items) - 1) + ['└── ']

    for pointer, item_name in zip(pointers, filtered_items):
        full_path = os.path.join(directory, item_name)
        print(prefix + pointer + item_name)

        if os.path.isdir(full_path):
            extension = '|   ' if pointer == '├── ' else '    '
            _generate_tree(full_path, prefix + extension, ignore_patterns)

if __name__ == "__main__":
    # --- 🖥️ HARDCODED PATHS ---
    # IMPORTANT: Replace these placeholder paths with the actual paths on your system.
    
    # 1. Path to the project directory you want to scan.
    #    Example for Windows: "C:/Users/YourUser/Documents/MyProject"
    #    Example for Mac/Linux: "/home/user/my_project"
    project_directory_path = "/home/rahul/dev/Learnings/agenticAIFramework"  # <-- EDIT THIS LINE
    
    # 2. Path to the ignore file. 
    #    Set this to the full path of your .gitignore file.
    #    If you set it to None, the script will auto-detect .gitignore in the root.
    ignore_file_path = "/home/rahul/dev/Learnings/.gitignore"  # <-- EDIT THIS LINE (or set to None)
    
    print(f"▶️ Running script with hardcoded project path: {project_directory_path}")
    display_directory_tree(project_directory_path, ignore_file_path)
```
